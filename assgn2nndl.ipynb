```python
from sympy import *
x=Symbol('x')
y=Symbol('y')
fun=x**2+y**2+10
parder=Derivative(fun,y)
dy=parder.doit()
parder=Derivative(fun,x)
dx=parder.doit()

x0=-1.0
y0=2.0
itr=100
n=0.1
for i in range(0,itr):
  x=symbols('x')
  y=symbols('y')
  dfxv=dx.subs(x,x0)
  dfxv=round(dfxv,2)
  dfyv=dy.subs(y,y0)
  dfyv=round(dfyv,2)
  dfx=(-1.0)*n*dfxv
  dfy=(-1.0)*n*dfyv
  x0=x0+dfx
  x0=round(x0,2)
  y0=y0+dfy
  y0=round(y0,2)
print(x0,y0)
print(f'min value obtained at x={x0},y={y0}')
```
